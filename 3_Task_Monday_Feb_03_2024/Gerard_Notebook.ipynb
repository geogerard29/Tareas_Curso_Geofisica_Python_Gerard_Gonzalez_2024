###5- Main Statistical Measures for the variable rainfall (Column Valor)

#### 5.1 Some quaintiles, mean, median and std for all the values.

```python
Df_Rain_10 = Df_Rain['Valor'].quantile(0.10)
print(f"The 25 % quaintile is {round(Df_Rain_10,2)}\n")

Df_Rain_50 = Df_Rain['Valor'].quantile(0.50)
print(f"The 50 % quaintile is {round(Df_Rain_50,2)}\n")

Df_Rain_90 = Df_Rain['Valor'].quantile(0.90)
print(f"The 90 % quaintile is {round(Df_Rain_90,2)}\n")

Df_Rain_Mean = Df_Rain['Valor'].mean()
print(f"The mean is {round(Df_Rain_Mean,2)}\n")

Df_Rain_Median = Df_Rain['Valor'].median()
print(f"The median is {round(Df_Rain_Median,2)}\n")

Df_Rain_Std = Df_Rain['Valor'].std()
print(f"The stantard desviation is {round(Df_Rain_Std,2)}\n")
```

#### 5.2 Some quaintiles, mean, median and std for the weather station CURITAL in January.

##### 5.2.1. filtering data

```python
# 1) A list with the month january for each year is created

Base_Year_January = '20CD-01-01 00:00'
List_Year_January = []
Index_C = 2    # Índice del carácter a modificar
Index_D = 3
for i in range(0,3):
  for j in range(0,10):
    Text_Definitive_January = Base_Year_January[:Index_C] + str(i) + str(j) + Base_Year_January[Index_D+1:]
    List_Year_January.append(Text_Definitive_January)
    if i == 2 and j == 3:
      break
List_Year_January
```

```python
# 2) A Dataframe with values of rainfall in january for each year is created
Df_Rain_January_CURITAL = Df_Rain[(Df_Rain['CodigoEstacion'] == 24030690) & (Df_Rain['Fecha'].isin(List_Year_January))]
Df_Rain_January_CURITAL
```

##### 5.2.2. Statistical Measures

```python
Df_Rain_January_CURITAL_10 = Df_Rain_January_CURITAL['Valor'].quantile(0.10)
print(f"The 25 % quaintile is {round(Df_Rain_January_CURITAL_10,2)}\n")

Df_Rain_January_CURITAL_50 = Df_Rain_January_CURITAL['Valor'].quantile(0.50)
print(f"The 50 % quaintile is {round(Df_Rain_January_CURITAL_50,2)}\n")

Df_Rain_January_CURITAL_90 = Df_Rain_January_CURITAL['Valor'].quantile(0.90)
print(f"The 90 % quaintile is {round(Df_Rain_January_CURITAL_90,2)}\n")

Df_Rain_January_CURITAL_Mean = Df_Rain_January_CURITAL['Valor'].mean()
print(f"The mean is {round(Df_Rain_January_CURITAL_Mean,2)}\n")

Df_Rain_January_CURITAL_Median = Df_Rain_January_CURITAL['Valor'].median()
print(f"The median is {round(Df_Rain_January_CURITAL_Median,2)}\n")

Df_Rain_January_CURITAL_Std = Df_Rain_January_CURITAL['Valor'].std()
print(f"The stantard desviation is {round(Df_Rain_January_CURITAL_Std,2)}\n")
```

###6- Gassuiana Distribution Graph

##**1-First Problem: Coordinates**

Create an algorithm in which the user types coordinates of latitude and longitude. After, print out " The coordinates are 62.22 N, 12.33 W " or something similar, the latitude and longitude must have two significant figures.


**Solution**

```python
Lat = float(input("Please type latitude (positive for N, negative for S): "))
Lon = float(input("Please type longitude (positive for E, negative for W): "))
Hem_Lat = None
Hem_Lon = None
if Lat > 0:
  Hem_Lat = "N"
else:
  Hem_Lat = "S"
if Lon > 0:
  Hem_Lon = "E"
else:
  Hem_Lon = "W"
print(f"coordinates of your point: {round(Lat,2)} {Hem_Lat}, {round(Lon,2)} {Hem_Lon}")
```

##**3-Third Problem: From ºC to ºF**

Create an algorithm in which the user type a value of temperature in degrees Celsius ºC and the program converts this value to degrees Fahrenheit ºF. After, it must print out a message that says "XXºC is equal to YYºF"


**Solution**

```python
Cel = float(input("Please type the value in ºC: "))
Fah = (Cel * (9/5)) + 32
print(f"{Cel} ºC is equal to {Fah} ºF")
```

## 4-Challenge: Prime number ##

This problem is more advanced than the previous ones and it is going to require independent research from what we saw in class. It consists of creating an algorithm in which the user types a number and it prints out if that number is or not a prime number.



```python
number = int(input("Please type a number: "))
remainder = None
theCounter = 2
for i in range(2,number):
  remainder = number % i
  if remainder == 0:
    print(f"{number} is not a prime number")
    break
  else:
    theCounter += 1
if theCounter == number:
    print(f"{number} is a prime number")
```

**Solution**

```python
import numpy as np

amountDices = int(input("Please type the amount of dices to throw: "))
throws = int(input("Please type the number of throws: "))

oneArrey = np.random.randint(1, 7, (amountDices * throws))
oneArrey = oneArrey.reshape(throws,amountDices)

for i in range(amountDices):
  print(f"\nThe values for the dice No {i+1}:")
  for j in range(oneArrey[:,i].size):
      print(f"Throw No {j+1}: {oneArrey[:,i][j]} ")
```

**Optional Challenge**

The algorithm must show to the user the most common number, the mean, the standard deviation, the minimum value, and the maximum value for each dice after 10 throws.



**Solution**

```python
print("DESCRIPTIVE STATISTICS VALUES")
for i in range(amountDices):
  vals, counts = np.unique(oneArrey[:,i], return_counts=True)
  mode_value = vals[counts == np.max(counts)]
  print(f"\nThe most common number is: {mode_value} ")
  print(f"The mean is: {oneArrey[:,i].mean()}\nThe standar deviation is: {oneArrey[:,i].std()} ")
  print(f"The minumun value is: {oneArrey[:,i].min()}\nThe maximun value is: {oneArrey[:,i].max()} ")
```
